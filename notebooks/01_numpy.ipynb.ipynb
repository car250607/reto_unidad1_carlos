En esta parte realizaremos todo lo relacionado con Numpy y sus utilidades en álgebra lineal para saber algunos datos que nos seran de ayuda para realizar el pipeline

#### 1. Extracción de datos del csv

```python
# Importamos las librerias que vamos a usar 
import numpy as np  
import pandas as pd

# pandas lee el archivo csv y lo convierte en una tabla
tabla1 = pd.read_csv('../data/ventas_techventas.csv', encoding='latin-1', parse_dates=['fecha'])
tabla1.head() # Primeras 5 filas 
```

#### 1.1 Creacion de arrays

```python
# Seleccionamos las columnas cantidad y precio_unitario por separado y con .to_numpy lasc onvertimos en arrays
cantidades = tabla1["cantidad"].to_numpy()
precios = tabla1["precio_unitario"].to_numpy()

# Prints para verificar las arrays creadas
print("Cantidades:", cantidades)
print("Precios:", precios)
```

### 1.2 Cálculo de ingresos brutos con Numpy (¿Error?)

```python
# Multiplicación elemento a elemento
ingresos_por_producto = cantidades * precios
print("Ingreso por producto:", ingresos_por_producto)

# Producto punto con np.dot (ingreso total)
ingreso_total_dot = np.dot(cantidades, precios)
print("Ingreso total (np.dot):", ingreso_total_dot)

# Lo mismo pero con el operador @
ingreso_total_arroba = cantidades @ precios
print("Ingreso total (@):", ingreso_total_arroba)

# Verificar que ambos dan el mismo resultado
print("¿Son iguales?", ingreso_total_dot == ingreso_total_arroba)
```

Nos podemos percatar de que hay un error en las operaciones, esto ocurre debido a que, si nos damos cuenta, en las arrays anteiormente creadas podemos ver valores que dicen "nan" lo que es un espacio vacio numericamente, o sea no hay nada escrito en esos espacios del csv donde aparece "nan".

Debido a esto, para completar con este punto de la actividad limpiaremos estos datos con ayuda de Numpy, generalmente se usa la libreria de Pandas pero veo necesario realizar esta pequeña limpieza para continuar.

### 1.2.1 Limpieza de nan

```python
# Eliminar filas donde cantidad o precio sean nan
mask = ~np.isnan(cantidades) & ~np.isnan(precios)
cantidades_limpias = cantidades[mask]
precios_limpios = precios[mask]

# Prints para observar la cantidad de valores que eliminamos con la limpieza de nan
print("Registros originales:", len(cantidades))
print("Registros limpios:", len(cantidades_limpias))
```

Con esto ya podremos realizar los cálculos de ingresos brutos de la empresa.

Quiero añadir que dejare el codigo anterior para poder visualizar el error que mencione y volvere a ejecutarlo despues de esta linea de limpieza.

### 1.2 Cálculo de ingresos brutos con Numpy (Corregido)

```python
# Multiplicación elemento a elemento
ingresos_por_producto = cantidades_limpias * precios_limpios
print("Ingreso por producto:", ingresos_por_producto)

# Producto punto con np.dot (ingreso total)
ingreso_total_dot = np.dot(cantidades_limpias, precios_limpios)
print("Ingreso total (np.dot):", ingreso_total_dot)

# Lo mismo pero con el operador @
ingreso_total_arroba = cantidades_limpias @ precios_limpios
print("Ingreso total (@):", ingreso_total_arroba)

# Verificar que ambos dan el mismo resultado
print("¿Son iguales?", ingreso_total_dot == ingreso_total_arroba)
```

Ahora si dan los resultados. 
### Conclusión
Tanto np.dot como el operador @ producen el mismo ingreso bruto total.
La diferencia con * es que este último devuelve un array con el ingreso 
por cada producto, mientras que @ devuelve un único valor total.

### 1.3 Estadísticas con Numpy

```python
# Extraer precio_unitario limpio
precios_stat = tabla1['precio_unitario'].dropna().to_numpy() # .dropna limpia con funcion de la libreria Pandas

# Estadísticas
media = np.mean(precios_stat)
mediana = np.median(precios_stat)
desviacion = np.std(precios_stat)
percentiles = np.percentile(precios_stat, [25, 75])

print("Media:", round(media, 3))                      # round nos ayuda a limitar la cantidad de decimales
print("Mediana:", round(mediana, 3))                   
print("Desviación estándar:", round(desviacion, 3))
print("Percentil 25:", percentiles[0])
print("Percentil 75:", percentiles[1])
```

Podemos observar con los resultados de este estadistico de los datos que el catalogo de TechVentas tiene una distribución de precios asimétrica, con la mayoria de los productos en rangos económicos (al rededor de $85-$95) pero con algunos productos un poco más costosos que elevan significativamente el promedio hasta $384

Quiero quizas dejar un análisis un poco más detallado de los datos optenidos.

### Media $384 vs Mediana $95
La media es mucho más alta que la mediana. Eso significa que hay pocos productos muy caros que jalan la media hacia arriba, pero la mayoría de productos son baratos.

### Desviación estándar $492
Es mayor que la media misma ($384). Eso significa que los precios están muy dispersos, no hay un precio "típico". El catálogo mezcla productos muy baratos con productos muy caros.

### Percentil 25: $85 y Percentil 75: $1200
El 25% de productos cuesta menos de $85 y el 75% cuesta menos de $1200. Esa diferencia tan grande confirma que el catálogo tiene dos rangos muy distintos: productos económicos y productos premium.
